```python
plt.figure(figsize=(14, 10))
sns.lineplot(data=df, x='Year', y='emissions_total_per_capita')
plt.title('Emission per year')
plt.xlabel('Years')
plt.ylabel('Emission')
plt.grid(True, which='both', linestyle='--', alpha=0.5)


#Key events| Ключевые события
plt.axvline(x = 1939, color = 'red', linestyle = '--', label = 'Start of WW2')
plt.axvline(x = 1945, color = 'green', linestyle = '--', label = 'End of WW2')
plt.axvline(x = 1949, color = 'blue', linestyle = '--', label = 'NATO was created')
plt.axvline(x = 1955, color = 'orange', linestyle = '--', label = 'Warsaw Pact Organization was created')
plt.axvline(x = 1991, color = 'yellow', linestyle = '--', label = 'End of ColdWar')

plt.legend()
plt.show()
```
